## Field

```python
class Validity(Enumeration):
    error = 0
    correct = 1
    forced = 2
    undefined = 3
```

```python
enum = Enum(bit_size=2, align_to=None, enumeration=None, byte_order='auto')
```

```python
enum = Enum(2, enumeration=Validity)
```
